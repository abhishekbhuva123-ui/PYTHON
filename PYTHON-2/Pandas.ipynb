```python
import pandas as pd

data = {
    'Department': ['IT', 'HR', 'IT', 'HR', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [50000, 40000, 60000, 45000, 70000]
}

df = pd.DataFrame(data)
```

```python
result=df.groupby('Department')['Salary'].agg(['sum','min','mean'])
print(result)
```

```python
import pandas as pd
import numpy as np

# 1. Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'A'],
    'Value': [10, 20, 10, 30, 40, 20, 10, 40],
    'Region': ['East', 'West', 'East', 'East', 'West', 'West', 'East', 'East']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("-" * 30)

# 2. Group by 'Category' and count the number of unique 'Value' entries in each group
# The result will be a pandas Series
unique_counts_series = df.groupby('Category')['Value'].nunique()

print("Unique counts per Category (pandas Series):")
print(unique_counts_series)
print("-" * 30)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "Maths": [80, 85, 90],
    "Science": [78, 82, 88],
    "English": [70, 75, 80]
})

result = df1.apply(lambda x:x.mean())
print(result)
```

```python
df3 = pd.DataFrame({
    "Sub1": [50, 60, 70],
    "Sub2": [55, 65, 75],
    "Sub3": [60, 70, 80]
})

result = df3.apply(lambda x: x.sum(), axis=1)
print(result)
```

```python
df5 = pd.DataFrame({
    "Marks1": [85, 70, 90],
    "Marks2": [80, 65, 95],
    "Marks3": [75, 60, 85]
})


def grade(total):
    if total >= 250:
        return "A"
    elif total >= 200:
        return "B"
    else:
        return "C"
    
df5['Total']=df5.apply(lambda x:x.sum(),axis=1)
df5['Grade']=df5["Total"].apply(grade)

print(df5)
```

```python
import pandas as pd
df7 = pd.DataFrame({
    "Marks": [80, 90, 85],
    "Bonus": [5, 10, 5],
    "Penalty": [2, 1, 3]
})

result = df7.apply(
    lambda x: [x.sum(), x.mean()],
    axis=1,
    result_type='expand'
)

print(result)
```

```python
df1 = pd.DataFrame({'A': [1, 2]})
df2 = pd.DataFrame({'B': [3, 4]})

result = pd.concat([df1, df2],axis=1)
print(result)
print(pd.concat([df1, df2], join='outer',axis=0))
#All columns from both DataFrames are included:outer
#Only common columns are included.:inner
print(pd.concat([df1, df2], join='inner'))
df1 = pd.DataFrame({'A': [1, 2]})
df2 = pd.DataFrame({'A': [3, 4]})
pd.concat([df1, df2], join='inner')
```
